```python
import pandas as pd

raw_data = pd.DataFrame({
        'bioguideId': ['A1', 'B2', 'C3'],
        'name': ['Edgar', 'Pedro', 'Alfonso'],
        'partyName': ['Democratic', 'Republican', 'Democratic'],
        'state': ['CA', 'TX', 'ny']
    })
```

```python
raw_data.head()
```

```python
silver_df = raw_data.copy()

silver_df = silver_df[silver_df['partyName'] == 'Democratic']
silver_df['state'] = silver_df['state'].str.lower()

silver_df.head()
```
